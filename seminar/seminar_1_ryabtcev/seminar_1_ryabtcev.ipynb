```python
# String

my_string = 'dlfmldmmdl;f'
print(my_string)
```

```python
# String Operstions

print(my_string * 2)
print(my_string + 'Innit')
print('m' in my_string)
print(my_string[3])
print(my_string[4:9])
```

```python
# List

a = 'is'
b = 'nice'
my_list = ['my', 'list', a, b]
my_list2 = [[4, 5, 6, 7, 9], [3, 4, 5, 6, 1]]
print(my_list)
print(my_list2)
```

```python
# Selecting List Elements

# Subset

print(my_list[1])
print(my_list[-3])
```

```python
#Slice

print(my_list[1:3])
print(my_list[1:])
print(my_list[:3])
print(my_list[:])
```

```python
# Subset Lists of Lists

print(my_list2[1][0])
print(my_list2[1][:2])
```

```python
# List Operations

print(my_list + my_list)
print(my_list * 2)
print(len(my_list2) > 2)
```

```python
# List Operation

my_list = ['my', 'list', a, b]
print(my_list.extend('!'))
print(my_list.pop(-1))
```

```python
# Libraries
# Import numpy

import numpy as np

# Selective import
# Numpy Arrys

my_list = [1, 2, 3, 4]

my_array = np.array(my_list)
print(my_array)

my_2darray = np.array([[1, 2, 3], [4, 5, 6]])
print(my_2darray)
```

```python
# Selecting Numpy Array Elements
# Subset

print(my_array[1])
```

```python
# Slice

my_array[0:2]

# Subset 2D Numpy arrays

print(my_2darray[:, 0])
```

```python
# Numpy Array Operations

print(my_array > 3)
print(my_array * 2)
print(my_array + np.array([5, 6, 7, 8]))
```

```python
# Numpy Array Functions

print(my_array.shape)
print(np.append(my_array, 5))
print(np.insert(my_array, 1, 5))
print(np.delete(my_array, [1]))
print(np.mean(my_array))
print(np.median(my_array))
print(np.corrcoef(my_array))
print(np.std(my_array))
print("")
```
