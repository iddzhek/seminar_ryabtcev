```python
# Creating Arrays
from numpy.doc import misc
import numpy as np
a = np.array([1, 2, 3])
b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float)
c = np.array([[(1.5, 2, 3,), (4, 5, 6)], [(3, 2, 1), (4, 5, 6)]], dtype=float)
print(a)
print(b)
print(c)
```

```python
# Initial Placeholders

print(np.zeros((3, 4)))
print(np.ones((2, 3, 4), dtype=np.int16))
d = np.arange(10, 25, 5)
print(d)
print(np.linspace(0, 2, 9))

print(np.full((2, 2), 7))
print(np.eye(2))
print(np.random.random((2, 2)))
print(np.empty((3, 2)))
```

```python
# Insperting Your Array

print(a.shape)
print(len(a))
print(b.ndim)
print(c.size)
print(b.dtype)
print(b.dtype.name)
print(b.astype(int))
```

```python
# Array Mathematics

print(a - b)
print(np.subtract(a, b))

print(a + b)
print(np.add(a, b))

print(a / b)
print(np.divide(a, b))

print(a * b)
print(np.multiply(a, b))

print(np.exp(b))
print(np.sqrt(b))
print(np.sin(b))
print(np.cos(b))
print(np.log(b))
```

```python
# Comparison
print(a == b)
print(a < 2)
print(np.array_equal(a, b))
```

```python
#Linear Algebra
#Creating Matrices
from scipy import linalg, sparse
A=np.array(np.random.random((2, 2)))
print(A)
B=np.asmatrix(b)
print(B)
C=np.mat(np.random.random((10, 5)))
print(C)
D=np.mat([[3, 4], [5, 6]])
print(D)
```

```python
#Basic Matrix Routines
#Creating Sparse Matrices
F=np.eye(3, k=1)
print(F)
G=np.mat(np.identity(2))
print(G)
C[C>0.5]=0
H=sparse.csr_matrix(C)
I=sparse.csr_matrix(D)
J=sparse.dok_matrix(A)
print(sparse.isspmatrix_csc(A))
print(H, I, J)
```

```python
#Inverse
print(linalg.inv(A))
#Transposition
print(A.T)
#Trace
print(np.trace(A))
#Norm
print(linalg.norm(A))
print(linalg.norm(A, 1))
print(linalg.norm(A, np.inf))
#Rank
print(np.linalg.matrix_rank(C))
#Determinant
print(linalg.det(A))
#Solving linear problems
print(linalg.solve(A, b))
E=np.mat(a).T
print(E)
print(linalg.lstsq(F, E))
#Generalized inverse
print(linalg.pinv(C))
print(linalg.pinv2(C))
```

```python
#Matrix Functions
#Addition
print(np.add(A, D))
#Subtraction
print(np.subtract(A, D))
#Division
print(np.divide(A, D))
#Multiplication
print(A @ D)
print(np.multiply(D, A))
print(np.dot(A, D))
print(np.inner(A, D))
print(np.outer(A, D))
print(np.tensordot(A, D))
print(np.kron(A, D))
#Exponential Functions
print(linalg.expm(A))
#Trigonometric Functions
print(linalg.sinm(D))
#Hyperbolic Trigonometric Functions
print(linalg.sinhm(D))
#Matrix Square Root
print(linalg.sqrtm(A))
#Arbitrary Funcrions
print(linalg.funm(A, lambda x: x*x))
```

```python
#Decompositios
#Eiganvalues and Eigenvectors
la, v=linalg.eig(A)
l1, l2=la
print(la)
print(v[:,0])
print(v[:,1])
print(linalg.eigvals(A))
#Signular Value Decomposition
U, s, Vh=linalg.svd(B)
print(U, s, Vh)
M, N=B.shape
print(M, N)
#
#LU Decomposition
print('')
P, L, U=linalg.lu(C)
print(P, L, U)
print('')
#Sparse Matrix Decompositions
la, v=sparse.linalg.eigs(F, 1)
print(la, v)
print(sparse.linalg.svds(H, 2))
```
